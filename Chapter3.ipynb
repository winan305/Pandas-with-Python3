```python
import pandas as pd

#1. 시리즈에 데이터를 주어 직접 만들어보기
series = pd.Series(['banana', 42])
print(series)

series = pd.Series(['Wes McKinney', 'Creator of Pandas'])
print(series)

series = pd.Series(['Wes McKinney', 'Creator of Pandas'], index=['Person', 'Who'])
print(series)
```

```python
# 데이터프레임에 데이터를 주어 직접 만들기
# 데이터 프레임은 딕셔너리를 DataFrame 클래스에 전달해야 한다.

# 데이터 딕셔너리 생성
data_dict = {
    'Name' : ['Rosaline Franklin', 'William Gosset'],
    'Occupation' : ['Chemist', 'Statistician'],
    'Born' : ['1920-07-25', '1876-06-13'],
    'Died' : ['1958-04-16', '1937-10-16'],
    'Age' : [37, 61]
}

# 데이터 프레임 생성
scientists = pd.DataFrame(data_dict)
print(scientists)

# 인덱스를 따로 전달하여 생성하는 방법
# 데이터 딕셔너리 생성
data_dict = {
    'Occupation' : ['Chemist', 'Statistician'],
    'Born' : ['1920-07-25', '1876-06-13'],
    'Died' : ['1958-04-16', '1937-10-16'],
    'Age' : [37, 61]
}
# 인덱스 및 컬럼 리스트 생성
index = ['Rosaline Franklin', 'William Gosset']
columns = ['Occupation', 'Born', 'Age', 'Died']

scientists = pd.DataFrame(data = data_dict,
                          index = index,
                          columns = columns)
print(scientists)

# 정렬된 딕셔너리를 활용하기
from collections import OrderedDict

data_dict = OrderedDict([
    ('Name', ['Rosaline Franklin', 'William Gosset']),
    ('Occupation', ['Chemist', 'Statistician']),
    ('Born', ['1920-07-25', '1876-06-13']),
    ('Died', ['1958-04-16', '1937-10-16']),
    ('Age', [37, 61])
])
scientists = pd.DataFrame(data_dict)
print(scientists)
```

```python
#2. 시리즈 다루기

# 데이터 프레임 생성
data_dict = {
    'Occupation' : ['Chemist', 'Statistician'],
    'Born' : ['1920-07-25', '1876-06-13'],
    'Died' : ['1958-04-16', '1937-10-16'],
    'Age' : [37, 61]
}
index = ['Rosaline Franklin', 'William Gosset']
columns = ['Occupation', 'Born', 'Age', 'Died']


scientists = pd.DataFrame(data = data_dict,
                          index = index,
                          columns = columns)
first_row = scientists.loc['William Gosset']

# 데이터프레임의 한 행을 Series로 봄.
# 판다스의 데이터를 구성하는 기본 단위가 Series 이다.
print(type(first_row))
print(first_row)
```

```python
# 시리즈의 index, values, keys 사용하기

# index와 values는 프로퍼티.
print(first_row.index)
print(first_row.values)

# keys는 메서드로 index와 동일.
print(first_row.keys())

# index 속성의 특정 값 가져오기. 아래 두 가지는 동일한 값을 가져옴.
print(first_row.index[0])
print(first_row.keys()[0])

# 시리즈의 기초 통계(mean, min, max, std) 메서드 사용하기
ages = scientists['Age']
print(ages)
print(ages.mean())
print(ages.min())
print(ages.max())
print(ages.std())

# 이 외의 시리즈 메서드
# append : 2개 이상의 시리즈 연결
# describe : 요약 통계량 계산
# drop_duplicates : 중복값 없는 시리즈 반환
# equals : 시리즈에 해당 값을 가진 요소 확인
# get_values : 시리즈 값 구하기(values 프로퍼티)
# isin : 시리즈에 값이 포함되어 있는지
# min : 최솟값 반환
# max : 최댓값 반환
# mean : 산술 평균 반환
# median : 중간값 반환
# replace : 특정 값을 가진 시리즈 값을 교체
# sample : 시리즈에서 임의의 값을 반환
# sort_values : 값을 정렬
# to_frame : 시리즈를 데이터프레임으로 변환
```

```python
#시리즈 응용하기

# csv 파일로부터 데이터 불러오기
scientists = pd.read_csv('./data/scientists.csv')
print(scientists.head())
```

```python
# 나이 열 가져오기
ages = scientists['Age']

# 산술 평균 함수 사용
print(ages.max())
print(ages.mean())

# 불린 추출 사용하기
print(ages[ages > ages.mean()])
# 아래의 결과에서 1,2,3,7 인덱스가 참이기에 위 코드의 결과가 나온다.
print(ages > ages.mean())

# 아래는 위 불린 추출과 같은 결과를 가져온다.
T = True
F = False
manual_bool_values = [T, T, F, F, T, T, F, T]
print(ages[manual_bool_values])
```

```python
# 시리즈와 브로드 캐스팅
# ages > ages.mean() 의 결괎값의 개수가 여러개인 이유
# 시리즈나 데이터 프레임에 있는 모든 데이터를 한 번에 연산하는 것을 브로드캐스팅 이라고 함.

# 벡터와 스칼라로 브로드 캐스팅 수행하기

# 벡터간의 연산
print(ages + ages)
print(ages * ages)

# 벡터와 스칼라의 연산
print(ages + 100)
print(ages * 2)

# 길이가 서로 다른 벡터 연산
# 인덱스가 일치한 곳만 연산하며 나머지는 누락값(NaN)으로 처리됨.
print(ages + pd.Series([1, 100]))

# 정렬 후 연산
rev_ages = ages.sort_index(ascending=False)
print(rev_ages)
# 같은 인덱스끼리 연산하기 때문에 결과가 동일하다.
print(ages + rev_ages)
```

```python
#4. 데이터프레임 다루기

# 불린 추출하기
print(scientists[scientists['Age'] > scientists['Age'].mean()])

# 브로드 캐스팅하기
print(scientists * 2)
```

```python
#5 시리즈와 데이터프레임의 데이터 처리하기

# Born과 Died 열의 차료형을 datatime으로 바꾸고 포맷팅한다.
born_datetime = pd.to_datetime(scientists['Born'], format='%Y-%m-%d')
print(born_datetime)

died_datetime = pd.to_datetime(scientists['Died'], format='%Y-%m-%d')
print(died_datetime)

# 데이터프레임에 새로운 열을 추가해본다.
scientists['born_dt'], scientists['died_dt'] = (born_datetime, died_datetime)
print(scientists.head())

# 시간을 계산해본다.
scientists['age_days_dt'] = (scientists['died_dt'] - scientists['born_dt'])
print(scientists)

# 데이터프레임의 데이터 섞기
import random
random.seed(42)
random.shuffle(scientists['Age'])
print(scientists['Age'])

# 데이터 프레임의 열 삭제하기
print(scientists.columns)
scientists_dropped = scientists.drop(['Age'], axis=1)
print(scientists_dropped.columns)
```

```python
#6.데이터 저장하고 불러오기

# 데이터 저장
names = scientists['Name']
names.to_pickle('./output/scientists_names_series.pickle')
scientists.to_pickle('./output/scientists_df.pickle')

# 데이터 불러오기
scientists_name_pickle = pd.read_pickle('./output/scientists_names_series.pickle')
scientists_from_pickle = pd.read_pickle('./output/scientists_df.pickle')

print(scientists_name_pickle)
print(scientists_from_pickle)

# csv, tsv로 저장하기
names.to_csv('./output/scientists_names_series.csv')
scientists.to_csv('./output/scientists_df.tsv', sep='\t')
```
